```python
import pandas as pd
```

```python
df = pd.read_csv(r"SICK\SICK.txt", sep="\t", encoding="utf-8")
```

```python
df['sentence'] = 'A: ' + df['sentence_A'] + '  B: ' + df['sentence_B']
```

```python
df['sentence'][0]
```

```python
df_new = df[['sentence', 'entailment_label', 'SemEval_set']]
```

```python
df_new
```

```python
label_map = {"CONTRADICTION": 0, "NEUTRAL": 1, "ENTAILMENT": 2}
```

```python
df_new['entailment_label'] = df_new['entailment_label'].map(label_map)
```

```python
df_new
```

```python
df_new["SemEval_set"].unique()
```

```python
df_new = df_new.rename(columns={'entailment_label': 'label'})
```

```python
train_df = df_new[df_new['SemEval_set'] == 'TRAIN'].sample(n=800, random_state=42)
val_df   = df_new[df_new['SemEval_set'] == 'TRIAL'].sample(n=100, random_state=42)
test_df  = df_new[df_new['SemEval_set'] == 'TEST'].sample(n=100, random_state=42)
```

```python
train_df.to_csv('train.csv', index=False)
val_df.to_csv('val.csv', index=False)
test_df.to_csv('test.csv', index=False)
```

```python
text_df = train_df[["sentence", "label"]]
text_df.to_csv("text.csv", index=False)
```
